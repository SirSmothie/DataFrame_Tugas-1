```python
import pandas as pd

# Baca file Excel
df_excel = pd.read_excel('jumlah_produksi_sampah_berdasarkan_kabupatenkota.xlsx', sheet_name='data')
df_excel
```

```python
# 1 Melakukan filter data
import pandas as pd

# Filter Data
df = df_excel[['nama_kabupaten_kota', 'jumlah_produksi_sampah', 'tahun']]
df = df.rename(columns ={
    'nama_kabupaten_kota' : 'Kabupaten/Kota',
    'jumlah_produksi_sampah' : 'Produksi Sampah (ton)',
    'tahun' : 'Tahun'
})
df.to_csv('soal1.csv', index=False)
df.to_excel('soal1.xlsx', index=False)
```

```python
# 2 Menghitung total produksi sampah di seluruh kabupaten/kota di jawa barat untuk tahun tertentu
# Memilih tahun tertentu
tahun_tertentu = 2015
total_sampah = 0
# Menghitung total produksi sampah menggunakan iterrows
for i, row in df.iterrows():
    if row['Tahun'] == tahun_tertentu:
        total_sampah += row['Produksi Sampah (ton)']

# Menampilkan hasil
df_soal2 = pd.DataFrame({'Tahun': [tahun_tertentu], 'Jumlah Sampah': [total_sampah]})
df_soal2.to_csv('soal2.csv', index=False)
df_soal2.to_excel('soal2.xlsx', index=False)
print(df_soal2)
```

```python
# 3 Jumlahkan data pertahun
import pandas as pd

jumlah_data_pertahun = {}

for i, row in df.iterrows():
    tahun_produksi = row['Tahun']
    
    if tahun_produksi not in jumlah_data_pertahun:
        jumlah_data_pertahun[tahun_produksi] = 0
    
    jumlah_data_pertahun[tahun_produksi] += 1

print(jumlah_data_pertahun)
df_soal3 = pd.DataFrame.from_dict(jumlah_data_pertahun, orient='index', columns=['Jumlah Data'])
df_soal3.to_csv('soal3.csv', index=False)
df_soal3.to_excel('soal3.xlsx', index_label='Tahun')
```

```python
# 4 Jumlahkan data per kota/kabupaten per tahun
import pandas as pd

jumlahData_perTahun_kabkota = {}

for i, row in df.iterrows():
    kabupaten_kota = row['Kabupaten/Kota']
    tahun = row['Tahun']

    if (kabupaten_kota, tahun) not in jumlahData_perTahun_kabkota:
        jumlahData_perTahun_kabkota[(kabupaten_kota, tahun)] = 0
    
    jumlahData_perTahun_kabkota[(kabupaten_kota, tahun)] += 1

print(jumlahData_perTahun_kabkota)
df_soal4 = pd.DataFrame.from_dict(jumlahData_perTahun_kabkota, orient='index', columns=['Jumlah Data'])
df_soal4.to_csv('soal4.csv', index=False)
df_soal4.to_excel('soal4.xlsx', index_label='Kabuten/Kota')
    
```

```python
# Gabungan code

import pandas as pd

# Baca file Excel
df_excel = pd.read_excel('jumlah_produksi_sampah_berdasarkan_kabupatenkota.xlsx', sheet_name='data')
df_excel

# Filter Data
df = df_excel[['nama_kabupaten_kota', 'jumlah_produksi_sampah', 'tahun']]
df

# Memilih tahun tertentu, misalnya tahun 2015
tahun_tertentu = 2015
df_tahun_tertentu = df[df['tahun'] == tahun_tertentu]

# Menghitung total produksi sampah (dengan mengabaikan NaN)
total = df_tahun_tertentu['jumlah_produksi_sampah'].sum()

# Menampilkan hasil
print(f"\nTotal produksi sampah di Jawa Barat pada tahun {tahun_tertentu}: {total:.2f} ton")

# Menampilkan DataFrame awal
print("DataFrame Awal:")
print(df)

# Menghitung total produksi sampah per tahun menggunakan iterrows
total_per_tahun = {}

for i, row in df.iterrows():
    tahun_produksi = row['tahun']
    jumlah_produksi = row['jumlah_produksi_sampah']
    
    if tahun_produksi in total_per_tahun:
        total_per_tahun[tahun_produksi] += jumlah_produksi
    else:
        total_per_tahun[tahun_produksi] = jumlah_produksi

# Menampilkan hasil
for tahun_produksi, total in sorted(total_per_tahun.items()):
    print(f"Total produksi sampah di Jawa Barat pada Tahun {tahun_produksi}: {total:.2f} ton")


total_sampah_kota_tahun = {}

# Menghitung total produksi sampah per kota/kabupaten per tahun menggunakan iterrows
for i, row in df.iterrows():
    nama_kabupaten_kota = row['nama_kabupaten_kota']
    tahun = row['tahun']
    jumlah_produksi_sampah = row['jumlah_produksi_sampah']
    
    # Membuat kunci gabungan untuk kota/kabupaten dan tahun
    kota_kabupaten_tahun = (nama_kabupaten_kota, tahun)
    
    # Menjumlahkan jumlah produksi sampah
    if kota_kabupaten_tahun in total_sampah_kota_tahun:
        total_sampah_kota_tahun[kota_kabupaten_tahun] += jumlah_produksi_sampah
    else:
        total_sampah_kota_tahun[kota_kabupaten_tahun] = jumlah_produksi_sampah

# Mengonversi hasil ke DataFrame untuk menampilkan hasil dengan lebih baik
result_df = pd.DataFrame(
    total_sampah_kota_tahun.items(), 
    columns=['kota_kabupaten_tahun', 'total_produksi']
)

# Memisahkan kolom 'kota_kabupaten_tahun' menjadi 'nama_kabupaten_kota' dan 'tahun'
result_df[['nama_kabupaten_kota', 'tahun']] = pd.DataFrame(result_df['kota_kabupaten_tahun'].tolist(), index=result_df.index)

# Menghapus kolom gabungan 'kota_kabupaten_tahun' karena sudah dipisah
result_df.drop(columns=['kota_kabupaten_tahun'], inplace=True)

# Menyusun ulang urutan kolom agar lebih informatif
result_df = result_df[['nama_kabupaten_kota', 'tahun', 'total_produksi']]

# Menampilkan hasil penjumlahan per kota/kabupaten per tahun
print("\ntotal produksi sampah per kota/kabupaten per tahun:")
print(result_df)

```
